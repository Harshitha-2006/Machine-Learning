```python
#simple linear regression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#loading into pandas
df=pd.read_csv("simplelr.csv")
df
```

```python
%matplotlib inline 
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='green')
```

```python
lr=linear_model.LinearRegression()
lr.fit(df[['area']],df.price)
```

```python
# Predicting the price of a house with 3300 sqft area
lr.predict([[3300]])
```

```python
lr.coef_
```
