```python
d=load_digits()
```

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(d.data,d.target)
```

```python
lr=LogisticRegression(max_iter=10000)
```

```python
lr.fit(x_train,y_train)
```

```python
lr.fit(x_test,y_test)
```

```python
plt.matshow(d.images[78])
```

```python
d.target[78]
```

```python
lr.predict([d.data[78]])
```

```python
lr.predict(d.data[0:3])
```

```python
#To identify exactly where the model makes incorrect predictions, we use a confusion matrix.
yp=lr.predict(x_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,yp)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
```
